```python
import numpy as np
import matplotlib.pyplot as plt

def amdahl_speedup(P, N):
    return 1 / ((1 - P) + (P / N))

# Define the number of processors
processors = np.array([10, 100, 1000, 10000, 100000])

# Define parallel fractions (P values) from slide 22
parallel_fractions = [0.50, 0.90, 0.95, 0.99]

# Plot the speedup for each parallel fraction
plt.figure(figsize=(8, 6))
for P in parallel_fractions:
    speedups = [amdahl_speedup(P, N) for N in processors]
    plt.plot(processors, speedups, marker='o', label=f'P = {P}')

plt.xlabel("Number of Processors (N)")
plt.ylabel("Speedup")
plt.title("Amdahl's Law: Speedup vs Number of Processors")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.xscale("log", base=2)  # Log scale for better visualization
plt.xticks(processors, labels=processors)
plt.show()
```
